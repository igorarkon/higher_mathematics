x1 = (10, 10, 10)
x2 = (0, 0, -10)
x1 + x2 = (10, 10, 10) + (0, 0, -10) = (10 + 0, 10 + 0, 10 + (-10)) = (10, 10, 0)

Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
```

```python
def vector_len(x1, x2, x3):
    return np.sqrt(x1**2 + x2**2 + x3**2)

vector_len(10, 10, 10)
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2 * X + 5 * Y
Z1 = 2 * X + 5 * Y + 10
ax.plot_wireframe(X, Y, Z, color='r')
ax.plot_wireframe(X, Y, Z1)
show()
```

# 3. Задание (в программе)

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
import random as rnd

def polar_to_decart(R, a):
    return R * cos(a), R * sin(a)
```

```python
for i in range(5):
    R = round(rnd.uniform(0, 100), 2)
    a = round(rnd.uniform(0, 360 * np.pi / 180), 2)
    print('R, a = (', R, ',', a, ') | x, y = (', round(polar_to_decart(R, a)[0], 2), ',',\
          round(polar_to_decart(R, a)[1], 2), ')')
```
